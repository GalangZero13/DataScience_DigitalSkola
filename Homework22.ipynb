## **Standard Statistics**

```python
tenure = data['tenure']
month_chg = data['MonthlyCharges']
data['TotalCharges'] = np.nan_to_num(data['TotalCharges'])
total_chg = pd.to_numeric(data['TotalCharges'], errors = 'coerce')
```

```python
# print(tenure)
ten_mean = np.average(tenure)
print("Tenure's mean is ", np.round(ten_mean,3))
ten_median = np.median(tenure)
print("Tenure's median is ", int(ten_median))
ten_modus = mode(tenure)
print("Tenure's modus is ", int(ten_modus))
ten_stdev = np.std(tenure)
print("Tenure's deviation standars is ", np.round(ten_stdev, 3))
```

```python
# print(month_chg)
mon_mean = np.average(month_chg)
print("Monthly Charge's mean is ", np.round(mon_mean,3))
mon_median = np.median(month_chg)
print("Monthly Charge's median is ", int(mon_median))
mon_modus = mode(month_chg)
print("Monthly Charge's modus is ", int(mon_modus))
mon_stdev = np.std(month_chg)
print("Monthly Charge's deviation standars is ", np.round(mon_stdev, 3))
```

```python
# print(total_chg)
tot_mean = np.nanmean(total_chg)
print("Total Charge's mean is ", np.round(tot_mean))
tot_median = np.nanmedian(total_chg)
print("Total Charge's median is ", int(tot_median))
tot_modus = mode(total_chg)
print("Total Charge's modus is ", int(tot_modus))
tot_stdev = np.std(total_chg)
print("Total Charge's deviation standars is ", np.round(tot_stdev, 3))
```
