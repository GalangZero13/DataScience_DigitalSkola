```python
x = data['math score']
y = data['reading score']
z = data['writing score']
```

### **Getting Pearson Coefficient using Pandas**

```python
#Find Correlation between math score and reading score
correlation_1 = x.corr(y)
#Find Correlation between math score and writing score
correlation_2 = x.corr(z)
#Find Correlation between reading score and writing score
correlation_3 = y.corr(z)
```

```python
print('Pearson value between math and reading score is    :', np.round(correlation_1,3))
print('Pearson value between math and writing score is    :', np.round(correlation_2,3))
print('Pearson value between reading and writing score is :', np.round(correlation_3,3))
```

```python
data.corr()
```

### **Getting Pearson Coefficient using Scipy**

```python
from scipy import stats
```

```python
#Find Correlation between math score and reading score
corr_1 = scipy.stats.pearsonr(x,y)[0]
#Find Correlation between math score and writing score
corr_2 = scipy.stats.pearsonr(x,z)[0]
#Find Correlation between reading score and writing score
corr_3 = scipy.stats.pearsonr(y,z)[0]
```

```python
print('Pearson value between math and reading score is    :', np.round(corr_1,3))
print('Pearson value between math and writing score is    :', np.round(corr_2,3))
print('Pearson value between reading and writing score is :', np.round(corr_3,3))
```
