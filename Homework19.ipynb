## **Task 1**
### Show table of survival rate for each gender based on class they took

```python
task1 = data1[['survived', 'gender', 'class']]
```

```python
task1
```

```python
results = task1.pivot_table(index = 'gender', columns = 'class',  aggfunc=np.average)
```

```python
results.head()
```

## **Task 2**
### Show table of survival number for each gender based on class they took

```python
task2 = task1.loc[(task1['survived']==1)]
task2
```

```python
results2 = task2.pivot_table(index = 'gender', columns = 'class', aggfunc = np.sum)
```
