```python
import requests
from datetime import datetime

api_key = '054862d9ae36e1f0d0b358ee1c092fde'
location = input("Enter the city name: ")

complete_api_link = "https://api.openweathermap.org/data/2.5/weather?q="+location+"&appid="+api_key

api_link = requests.get(complete_api_link)
api_data = api_link.json()

date_time = datetime.now().strftime("%d %b %Y | %I:%M:%S %p")
min_temp = ((api_data['main']['temp_min']-273.15))
max_temp = ((api_data['main']['temp_max']-273.15))
pressure = api_data['main']['pressure']
visibility = api_data['visibility']
weather_desc = api_data['weather'][0]['description']

print("--------------------------------------------")
print("Weather Stats for - {} || {}".format(location.upper(), date_time))
print("--------------------------------------------")
print('Minimum Temperature is:', min_temp, 'degree celcius')
print('Maximum Temperature is:', max_temp, 'degree celcius')
print('Weather Pressure is   :', pressure, 'pascals')
print('Weather Visibility is :', visibility, 'meters')
print("Current Weather Desc.  :", weather_desc)
```
