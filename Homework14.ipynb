```python
import numpy as np
base = np.array([10,10,20,20,30,30,40,10,15,15,12,20,30])
print(base)
```

### **Distinct Values**

```python
distinct = np.unique(base)
print('Value yang unik dari array:')
print(distinct)
```

### **Index from Distinct Values**

```python
distinct, index = np.unique(base, return_index=True)
print(index)
```

### **Sub-Array Division**

```python
split = np.split(distinct, [3])
print(split)
```
