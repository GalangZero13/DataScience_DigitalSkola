## **Soal No.1**

```python
masukkan = int(input('Masukkan sebuah angka : '))
i = 0
while i < masukkan:
  for j in range(i+1):
    print(j+1, end='')
  print('')
  i+=1
```

## **Soal No.2**

```python
angka = int(input('Masukkan angka positif : '))
sisa = 0

while angka>0:
  sisa = ((sisa*10)+(angka%10))
  angka = int(angka/10)

print(sisa)
```
