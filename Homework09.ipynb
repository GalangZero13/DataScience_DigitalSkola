### **HARD**

```python
X = input("Masukkan warna (putih, hitam, merah, hijau)? ")
putih = 5
hitam = 12
merah = 8
hijau = 5
probability = 0.0
total = putih + hitam + merah + hijau

if X == 'putih' or X == 'Putih' or X == 'PUTIH':
  probability = putih/total
  print("Peluangnya adalah ", probability)
elif X == 'hitam' or X == 'Hitam' or X == 'HITAM':
  probability = hitam/total
  print("Peluangnya adalah ", probability)
elif X == 'merah' or X == 'Merah' or X == 'MERAH':
  probability = merah/total
  print("Peluangnya adalah ", probability)
elif X == 'hijau' or X == 'Hijau' or X == 'HIJAU':
  probability = merah/total
  print("Peluangnya adalah ", probability)
else:
  print(0)
```
