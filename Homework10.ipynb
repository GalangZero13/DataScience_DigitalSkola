# **Menentukan tarif perjalanan**

image.png

```python
jarak = int(input('Masukkan jarak :'))
tarif = 10000
if jarak>1:
  if 1<jarak<=10:
    tarif += (jarak*2500)
    print(tarif)
  elif 10<jarak<=25:
    tarif += 10*2500
    tarif += ((jarak-10)*2250)
    print(tarif)
  elif 25<jarak<=50:
    tarif += 10*2500
    tarif += 15*2250
    tarif += ((jarak-25)*2000)
    print(tarif)
  elif 50<jarak<=500:
    tarif += 10*2500
    tarif += 15*2250
    tarif += 25*2000
    tarif += ((jarak-50)*1750)
    print(tarif)
  else:
    print('Tidak bisa melakukan perjalanan, batas maksimal 500 km')
else:
  print(tarif)
```

# **Mencari urutan kedua dari tiga angka.**

```python
print('Masukkan 3 angka secara urutan acak')
print('Batas angka adalah 5 juta')
a = int(input('Angka pertama : '))
b = int(input('Angka kedua   : '))
c = int(input('Angka ketiga  : '))
batas = 5000000

if (a>batas) or (b>batas) or (c>batas):
  print("Bilangan di luar jangkauan")
else:
  if (a<=b<=c) or (c<=b<=a):
    print("Bilangan tengahnya adalah ", b)
  elif (a<=c<=b) or (b<=c<=a):
    print("Bilangan tengahnya adalah ", c)
  elif (b<=a<=c) or (c<=a<=b):
    print("Bilangan tengahnya adalah ", a)
  else:
    print("Tidak boleh ada bilangan ganda")
  
```
