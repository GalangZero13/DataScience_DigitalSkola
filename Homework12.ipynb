## **No.2**: Statistical Problem

```python
import math
list1 = ['2', '3', '4', '5', '6']
for i in range(len(list1)):
  list1[i] = int(list1[i])
print(list1)
```

```python
average = 0
variance = 0
std_dev = 0
size_of_list = len(list1)

sum_of_all = 0
for i in list1:
  sum_of_all += i

average = sum_of_all/size_of_list
print('average : ', average)

delta = 0
total_diff_sqrd = 0
for i in list1:
  delta = i-average
  total_diff_sqrd += delta**2

variance = total_diff_sqrd/size_of_list
print('variance : ', variance)

std_dev = math.sqrt(variance)
print('standard deviation : ', std_dev)

```
