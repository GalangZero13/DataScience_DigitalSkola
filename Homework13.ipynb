```python
class Preparation:
  def __init__(self, DataInput):
    self.input = DataInput
  
  def process(self):
    output = []

    for data in self.input:
      traffic = 0
      for devices in data['device']:
        name_devices = list(devices.keys())[0]
        visit_devices = list(devices.values())[0]
        traffic += visit_devices
      output.append(({'date':data['date'], 'traffic':traffic}))
    
    self.output = output
    return self.output
```
