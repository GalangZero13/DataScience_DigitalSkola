# These data are the records of Winter Olympics Top Participant Medalist from 1924 until 2018,
# In it we could see the top countries medal owner along with their specific medals that were held as well as their participating hosts country and the city it was held throughout its history.

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../input/winter-olympic-medals-1924-2018/Winter_Olympic_Medals.csv')
data
```

```python
print('The Data Type for each Column in the dataset is as follows :')
data_type = data.info(verbose=True)
```

```python
print('-----------BASIC STATISTICS-----------')

print('The most frequent appearing top contestant countries:')
country_name = data[['Country_Name']]
print(country_name.mode(),'\n')

print('Total Gold Medals')
gold_medals = data[['Gold']]
print(gold_medals.sum(), '\n')

print('Total SIlver Medals: ')
silver_medals = data[['Silver']]
print(silver_medals.sum(), '\n')

print('Total Bronze Medals: ')
bronze_medals = data[['Bronze']]
print(bronze_medals.sum(), '\n')

unique = data['Country_Name'].unique()
print('ALL OF THE TOP PARTICIPANT THROUGHOUT YEARS ARE: ')
for i in unique:
    print(i)
    
print('\nThe most frequent host country')
host_name = data[['Host_country']]
print(host_name.mode())
```
